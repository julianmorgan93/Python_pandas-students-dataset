## Filtering Dataframe


```python
# Return students with mark geater than 90
df["mark"]>90
# This metthod is known as boolean masking
```

```python
df[df["mark"]>90]
#This will show the actual students with marks greter than 90
```

- Return students with mark less than 50

```python
df[df["mark"]<50]
```

- Return first 5 female students.

```python
df[df["gender"]=="female"].head()
# we need to use == not = ## == means to compare & match
```

- Return students at class Four



```python
df[df["class"]=="Four"]
```

- Return male students at class Four

```python
df[(df["gender"]=="male") & (df["class"]=="Four")]
# each condition needs to be in () the DFs inside needs to be ()
```

- Return IDs greater than 20 but less than 30

```python
df[(df["id"]>20) & (df["id"]<30)]
```

- Return students in class Four using loc operator

```python
df.loc[df["class"]=="Four", :]
## because we are filtering rows, criteria written first.
# : means return all columns
```

```python
df.iloc[df["class"]=="Four", :]
# We cannot use iloc for boolean masking!
```

- Return male students in class Four

```python
df.loc[(df["class"]=="Four") & (df["gender"]=="male"), ["name","class","gender"]]
```

- Return students at class Six or class Five

```python
df[(df["class"]=="Five") | (df["class"]=="Six")].sample(5)
 # | symbol is OR in python # , sample(5) will pick 5 random students in class Six OR Five
```

## How to sort results

```python
df.sort_values(by="mark")
# can use any column to sort by
# default sorting is ascending
```

```python
df.sort_values(by="mark", ascending=False)
#Python doesn't have Descending sort so we need to use ascending=False
```

```python
df.loc[df["gender"]=="male", :].sort_values(by="name").head()
# In this case we filter by gender and sort values by name
```

```python
# return student who are female or at class seven
df[(df["gender"]=="female") | (df["class"]=="Seven")].sort_values(by="mark")
```

```python
# Write code to find information of female students who are above the classroom average. Sort the results in descending order.
df[(df["gender"]=="female")&(df["mark"]>df["mark"].mean())].sort_values(by="mark",ascending=False)

# 1st step: - filter by gender: df[(df["gender"]=="female")
# 2nd step: filter by class average &(df["mark"]>df["mark"].mean())]
# 3rd step: sort the values as descending .sort_values(by="mark",ascending=False)

```

```python
df["mark"].mean()
```

```python
average = df["mark"].mean()
# creating average variable , df.loc version

df.loc[(df["gender"]=="female") & (df["mark"]>average) , ["name","mark", "gender"]].sort_values(by="mark", ascending=False)
```

```python
df.loc[(df["gender"]=="female") & (df["mark"] > df["mark"].mean()), ["name","mark", "gender"]].sort_values(by="mark", ascending=False)

# df.loc version without average variable.
```

```python
# Assign dataframe named `df` out of the following list of dictionaries
items = [
    {"item name": "USB cable", "price": "$10.99", "type": "USB C to USB C"},
    {"item name": "USB cable", "price": "$10.99", "type": "USB A to USB C"},
    {"item name": "Batteries", "price": "$9.99", "type": "AA"},
    {"item name": "Batteries", "price": "$8.99", "type": "AAA"},
    {"item name": "Mouse", "price": "$12.99", "type": "Wireless USB"},
]

# Create a new DataFrame from the items list
df_items = pd.DataFrame(items)

# Convert the price column into a float data type. First replace $ and then convert to float.
df_items["price"] = df_items["price"].str.replace("$", "").astype(float)

# Display the updated DataFrame
print(df_items)
```

```python
#Create a new column named is_automatic that holds a Boolean if that given vehicle is an automatic transmission or not.
df_items["is_automatic"] = df_items["type"].str.contains("Automatic")
```
